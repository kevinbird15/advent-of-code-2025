```
space_checks = [(dr, dc) for dr in [-1, 0, 1] for dc in [-1, 0, 1]]
```

```
space_checks.remove((0,0))
```

```
num_moveable_rolls=0
for i,si in enumerate(sample_input):
    for j, val in enumerate(si):
        if sample_input[i][j]!='@':
            continue
        num_rolls=0
        for check_row, check_col in space_checks:
            if i+check_row >=0 and j+check_col >= 0:
                try:
                    val = sample_input[i+check_row][j+check_col]
                except IndexError: 
                    continue
                if val == '@': 
                    num_rolls+=1
        if num_rolls <= 3:
            print(f"{i},{j} - {num_rolls}")
            num_moveable_rolls+=1
print(num_moveable_rolls)
```

The issue is that Python strings are **immutable** - you can't change individual characters in a string directly.

You have a few options to work around this:

1. **Convert each row to a list of characters** (so `sample_input` becomes a list of lists)
2. **Rebuild the string** using slicing: `sample_input[i] = sample_input[i][:j] + '.' + sample_input[i][j+1:]`
3. **Use a 2D numpy array** of characters

Which approach would you like to try? The first option (list of lists) is probably the most straightforward for this kind of grid manipulation.
