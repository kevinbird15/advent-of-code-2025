### Part 2

```
import math
def dbid_is_invalid(dbid_str:str) -> bool:
    "Determine whether a dbid is valid or not"
    dbid_len = len(dbid_str)
    if dbid_len==1: 
        return False
    for i in range(1, math.ceil(dbid_len/2)+1): 
        if len(set([dbid_str[j:j+i] for j in range(0, dbid_len, i)])) == 1:
            return True
    return False
```
