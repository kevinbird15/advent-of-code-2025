Yes! There are a few ways to multiply a list of numbers together:

1. **Using `math.prod()`** (Python 3.8+) - this is the direct equivalent to `sum()`:
   ```python
   import math
   math.prod([2, 3, 4])  # Returns 24
   ```

2. **Using `reduce()` with `operator.mul`**:
   ```python
   from functools import reduce
   from operator import mul
   reduce(mul, [2, 3, 4])  # Returns 24
   ```

3. **Using `numpy.prod()`** (if you're already using numpy):
   ```python
   import numpy as np
   np.prod([2, 3, 4])  # Returns 24
   ```

The first option (`math.prod()`) is probably the simplest and most similar to `sum()`.

````
numbers = sample_input[:-1]
signs = sample_input[-1]
````

````
from collections import defaultdict
col_nums = defaultdict(str)
for row in numbers:
    for i,space in enumerate(row):
        col_nums[i]+=space
sign_list = signs.split()
sign_list.reverse()
s = sign_list.pop()
eq_result=0 if s=='+' else 1
total_result = 0
for _,num in col_nums.items():
    try:
        if s == '+': 
            eq_result+=int(num)
        elif s == '*':
            eq_result*=int(num)
        else: 
            raise
        # print(num, eq_result) 
    except ValueError: # Split between equations 
        total_result+=eq_result
        s = sign_list.pop()
        eq_result=0 if s=='+' else 1
total_result+=eq_result
print(total_result)
````
